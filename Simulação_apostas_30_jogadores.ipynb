```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def simular_jogador(n):
    saldo = 0
    historico = []
    for _ in range(n):
        if np.random.rand() < 0.3:
            saldo += 20
        else:
            saldo -= 10
        historico.append(saldo)
    return historico

resultados_30 = [simular_jogador(1000) for _ in range(30)]

plt.figure(figsize=(12, 6.75))
for saldo in resultados_30:
    plt.plot(saldo, alpha=0.5)
plt.axhline(0, color='black', linestyle='--')
plt.title("Imagem 2 - Simulao: 30 Jogadores, 1000 Apostas Cada")
plt.xlabel("Nmero de Apostas")
plt.ylabel("Saldo (R$)")
plt.grid(True)
plt.tight_layout()
plt.show()
```
